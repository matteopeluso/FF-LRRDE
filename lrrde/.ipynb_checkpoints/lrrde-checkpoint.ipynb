```python
## Matteo Peluso Marzo 2019
import os
import time


import seaborn
seaborn.set(style='whitegrid')
seaborn.set_context('paper')

path_data = "/home/matteo/Scrivania/silver"
os.chdir(path_data)

# Ensemble of function
fun_path = "./utils/"
os.chdir(fun_path)

from parse_input     import *
from distances       import *
from multi_objective import *
from normal_solver   import *
from testing         import *
from out_function    import *
from iterative_regression import *


# --------------------------- User ------------------------------#
## numero di configurazioni totale pari a 120

input_params = {}
flag_files   = {}

input_params['n_conf']        = 239
input_params['n_test']        = 29   # dimension test set
input_params['n_train']       = 210  # dimension training set
input_params['len_vs']        = 0.09
input_params['n_vs']          = 2

input_params['index_ion']     = 1
input_params['N_atom']        = [385, 97]
input_params['N_water']       = [128, 32]

input_params['old_params']    = [1, 0.278,0.4]   # Q, sigma, eps :: Spezia

input_params['n_functions']   = 7 # 0 --> Parametri Ag: c6,c12   VS: q
                                   # 1 --> Parametri Ag: c6,c12   VS: q, c12
                                   # 2 --> Parametri Ag: c6,c12   VS: q, c6,c12
                                   # 3 --> Parametri Ag: q,c6,c12 VS: q,c6,c12
                                   # 4 --> Parametri Ag: q,c6,c12 VS: q 
                                   # 5 --> Parametri Ag: q,c6,c12 VS: q , c12
                                   # 6 --> Parametri bonus q,c6,c12 VS: q,c6,c12
                                   # 7 --> Parametri bonus c6,c12 VS: q c6
                
input_params['q_value']       = 0.0# if nfunctions = 0,1,2 inserire valore della carica
input_params['file_dir']      = "/home/matteo/Scrivania/silver/out_cn2_extended"
input_params['outdir']        = "/home/matteo/Scrivania/silver/mod_func"
input_params['counterpoise']  = True
input_params['borders']       = np.matrix([10e-15, 10e-1])

flag_files['topology']        = True
flag_files['energy']          = True
flag_files['force']           = True  
flag_files['coordinate']      = True 

# ---------------------------------------------------

## Parsing + Distance matrices
t0 = time.time()
_parser = parser(input_params, flag_files)
_parser.eval_distances()
input_params = _parser.input_params

# Training set
energy_ts = input_setup(input_params, flag="energy", weight=[1,1])
force_ts  = input_setup(input_params, flag="force",  weight=[1,1])
data_set  = training_set(energy_ts, force_ts )

data_set.model_descriptor()
t1 = time.time()
total = t1-t0
print("time building Training Set {}".format(total))
```

```python
# ------------------------ Fitting -------------------------------------#
t0 = time.time()
print("Evaluating parameters")
print("-------")
lrrde = eval_lrrde(data_set)
lrrde.eval_de()
# ------------------------ Testing -------------------------------------#
t1 = time.time()
total = t1-t0
print("time fitting {}".format(total))
#
print("-------")
print("Testing model")
print("-------")
#
testing_data(data_set = data_set, params = lrrde.params, plot = 'y')
```
